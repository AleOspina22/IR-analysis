### Rank Biased Overlap (RBO)

```python
# https://towardsdatascience.com/how-to-objectively-compare-two-ranked-lists-in-python-b3d74e236f6a


def rbo(S,T, p= 0.9):
    """ Takes two lists S and T of any lengths and gives out the RBO Score
    Parameters
    ----------
    S, T : Lists (str, integers)
    p : Weight parameter
    
    Returns
    -------
    Float of RBO score
    """
    
    # Fixed Terms
    k = max(len(S), len(T))
    x_k = len(set(S).intersection(set(T)))
    
    summation_term = 0

    # Loop for summation
    # k+1 for the loop to reach the last element (at k) in the bigger list    
    for d in range (1, k+1): 
            # Create sets from the lists
            set1 = set(S[:d]) if d < len(S) else set(S)
            set2 = set(T[:d]) if d < len(T) else set(T)
            
            # Intersection at depth d
            x_d = len(set1.intersection(set2))

            # Agreement at depth d
            a_d = x_d/d   
            
            # Summation
            summation_term = summation_term + math.pow(p, d) * a_d

    # Rank Biased Overlap - extrapolated
    rbo_ext = (x_k/k) * math.pow(p, k) + ((1-p)/p * summation_term)

    return rbo_ext
```

```python
initial_correlation = rbo(S,T) # without using the weight
print("The RBO score means that the two lists are nearly similar in", initial_correlation)

# as the result is 0.76, there is a good correlation but there aren't the same time of document retrieved for the same query 
```

```python
# define the weight (Wrbo) calculator function to help us choose p. 
# p for giving a weightage (Wrbo) to the top d elements under comparison
# d is the depth into the list ( if d = 5 , means first 5 documents)


def weightage_calculator(p,d):
    

    summation_term = 0

    for i in range (1, d): # taking d here will loop upto the value d-1 
        summation_term = summation_term + math.pow(p,i)/i


    Wrbo_1_d = 1 - math.pow(p, d-1) + (((1-p)/p) * d *(np.log(1/(1-p)) - summation_term))

    return Wrbo_1_d
```

```python
# the first 5 documents are normally the most important for the user. 
# for that I give more weight p = 0.8
# the result will later be implement in the rbo (S,T,P) where now it's possible to do again the comparison between
#    both list but given more weight (importance) to the first 5 document in the list

p = 0.8
d = 5 

weightage_calculator(p,d)

# If I take the first 5 document, the result is 0.67, means that the correlation between this first 5 documents is low 
# If I take the first 10, the result will be better (0.85)
# Means, that, when I take into consideration more 

```

# How much documents of the second query are in the third query, how much of the third query are in the fourth query, etc

```python
def get_query_intersections(df1, query_ids):
    query_data = []
    for query_id in query_ids:
        query_data.append(set(df1[df1['qID'] == query_id]['docID'].tolist()))
        
    intersection_lengths = []
    for i in range(len(query_data) - 1):
        intersection = query_data[i].intersection(query_data[i+1])
        intersection_lengths.append(len(intersection))
        
    return intersection_lengths
```

```python
query_ids = ["132_2-1", "132_2-3", "132_2-5", "132_2-7", "132_2-9"]
intersection_lengths = get_query_intersections(df1, query_ids)
print(intersection_lengths)
```

## EFFECTIVENESS 

```python
# effectiveness of the first question with respect to the last one in whiche the user is satisfied with the anser given.
# The function get_top_10_documents takes in the dataframe df1 and a list of query IDs query_ids. 
# It first sorts df1 by query ID and score, groups the sorted dataframe by query ID, selects the first 10 documents 
# for each group, and stores the resulting sets of top 10 documents in a new dataframe top_10_documents.

#The function then computes the number of shared documents between each pair of consecutive queries by 
# extracting the sets of top 10 documents for each query pair from top_10_documents, taking the intersection of 
# the two sets, and computing the length of the resulting set.

def get_top_10_documents(df1, query_ids):
    # sort the dataframe by query ID and score
    df1_sorted = df1.sort_values(['qID', 'score'], ascending=[True, False])
    
    # group the sorted dataframe by query ID and select the first 10 documents for each group
    top_10_documents = (df1_sorted.groupby('qID')
                        .head(10)
                        .groupby('qID')['docID']
                        .apply(set)
                        .reset_index(name='top_10_docs'))

    # compute the number of shared documents between each pair of consecutive queries
    shared_docs_counts = []
    for i in range(len(query_ids) - 1):
        query_pair = [query_ids[i], query_ids[i+1]]
        query_pair_docs = top_10_documents[top_10_documents['qID'].isin(query_pair)]['top_10_docs']
        shared_docs = query_pair_docs.iloc[0].intersection(query_pair_docs.iloc[1])
        shared_docs_counts.append(len(shared_docs))
    
    return shared_docs_counts
```

```python
query_ids = ["132_2-1", "132_2-3", "132_2-5", "132_2-7", "132_2-9"]
shared_docs_counts = get_top_10_documents(df1, query_ids)
print(shared_docs_counts)
```

## PERFORMANCE OF THE QUERY

### precision, recall and f1 score

```python
def evaluate_query_performance(df1, query_id):
    # get the set of relevant documents for the query
    relevant_docs = set(df1[(df1['qID'] == query_id) & (df1['tag'] == '1')]['docID'].tolist())

    # get the set of retrieved documents for the query
    retrieved_docs = set(df1[df1['qID'] == query_id]['docID'].tolist())

    # compute precision, recall, and F1 score
    precision = len(relevant_docs.intersection(retrieved_docs)) / len(retrieved_docs) if len(retrieved_docs) > 0 else 0
    recall = len(relevant_docs.intersection(retrieved_docs)) / len(relevant_docs) if len(relevant_docs) > 0 else 0
    f1_score = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0

    # print the results
    print(f'Query {query_id}:')
    print(f'Precision: {precision:.2f}')
    print(f'Recall: {recall:.2f}')
    print(f'F1 Score: {f1_score:.2f}')
```

```python
query_ids = df1['qID'].unique().tolist()
for query_id in query_ids:
    evaluate_query_performance(df1, query_id)

```

### MAP and NCDP

```python
import numpy as np
import pandas as pd
from sklearn.metrics import average_precision_score, ndcg_score

def evaluate_results(df1):
    # Group by query ID
    grouped = df1.groupby('qID')
    
    # Initialize lists to store MAP and NDCG values for each query ID
    map_scores = []
    ndcg_scores = []
    
    # Iterate over query IDs
    for qid, group in grouped:
        # Extract relevant scores and ranks
        scores = group['score'].values
        ranks = group['rank'].values
        
        # Calculate MAP
        labels = np.zeros_like(scores)
        labels[scores > 0] = 1
        ap_score = average_precision_score(labels, -ranks)
        map_scores.append(ap_score)
        
        # Calculate NDCG
        ndcg = ndcg_score([labels], [scores], k=len(scores))
        ndcg_scores.append(ndcg)
    
    # Return mean MAP and NDCG scores
    return np.mean(map_scores), np.mean(ndcg_scores)
```

```python
map_score, ndcg_score = evaluate_results(df1)
print("Mean Average Precision (MAP): {:.4f}".format(map_score))
print("Normalized Discounted Cumulative Gain (NDCG): {:.4f}".format(ndcg_score))
```

#### A MAP (Mean Average Precision) score of 1.0 means that all relevant documents were retrieved at the top of the search results for all queries. This is a perfect score, indicating that the retrieval system is performing optimally.

Similarly, an NDCG (Normalized Discounted Cumulative Gain) score of 1.0 means that all the relevant documents were retrieved at the top of the search results for all queries, and their relevance was correctly ordered. This is also a perfect score, indicating optimal performance.

In practice, it is difficult to achieve perfect scores, and the actual scores achieved by a retrieval system will depend on factors such as the quality of the search algorithm, the relevance judgments used to evaluate the system, and the nature of the data being searched.
